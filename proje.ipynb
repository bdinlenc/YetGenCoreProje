```python
#hata var 
import pandas as pd

# Verileri yükleme
df = pd.read_csv('netflix_titles.csv')

# Sadece film ve TV şovlarını içeren veri setleri oluşturma
df_movies = df[df['type'] == 'Movie']
df_tv_shows = df[df['type'] == 'TV Show']

# Yıla göre gruplayıp toplam sayıları hesaplama
movies_per_year = df_movies.groupby('release_year').size()
tv_shows_per_year = df_tv_shows.groupby('release_year').size()

# En yüksek puanlı filmi bulma ve adını getirme
highest_rated_movies = df_movies.loc[df_movies.groupby('release_year')['rating'].idxmax()][['release_year', 'title']]
highest_rated_movies.set_index('release_year', inplace=True)

# Verileri birleştirme
result = pd.concat([movies_per_year, tv_shows_per_year, highest_rated_movies], axis=1)
result.columns = ['Movies', 'TV Shows', 'Highest Rated Movie']

# Grafik oluşturma
result.plot(kind='bar', figsize=(12,6), color=['green', 'blue', 'purple'])
```

```python

#NEDEN EMPTY DATA FRAEME

import pandas as pd

# Veri setini oku
df = pd.read_csv('netflix_titles.csv')

# Rating verilerini string olarak tanımla
df['rating'] = df['rating'].astype(str)

# Fonksiyonu tanımla
def recommend_content(data, content_type, genre, country):
    # İstenilen kriterlere göre filtreleme yap
    filtered_data = data[(data['type'] == content_type) & 
                         (data['listed_in'].str.contains(genre)) & 
                         (data['country'].str.contains(country))]

    # Rating verilerini string olarak tanımladığımızdan dolayı burada string metotlarını kullanabiliriz
    sorted_data = filtered_data.sort_values('rating', ascending=False)

    # İlk 5 öneriyi döndür
    return sorted_data[['title', 'rating']].head(5)

# Fonksiyonu kullanarak önerileri al ve ekrana yazdır
recommendations = recommend_content(df, content_type='TV Show', genre='Documentaries', country='Turkey')
print(recommendations)


```

```python


#TOPLANTIDA BAKILSIN
import pandas as pd

# Verileri yükleme
df = pd.read_csv('netflix_titles.csv')

# Sadece film verilerini içeren veri seti oluşturma
df_movies = df[df['type'] == 'Movie']

# Yıla göre gruplayıp toplam sayıları hesaplama
movies_per_year = df_movies.groupby('release_year').size()

# En yüksek puanlı filmi bulma ve adını getirme
def get_most_common_genres(df):
    genres = df['listed_in'].str.split(', ')
    return pd.Series([x for row in genres for x in row]).value_counts()

most_common_genres = get_most_common_genres(df_movies) #burada amaç her turun kaç defa geçtiğini hesaplayıp dizi olarak döndürmek
most_common_genres = most_common_genres[most_common_genres >= most_common_genres.max()] #en cok tekrar eden turler burada saklanır
highest_rated_movie = df_movies[df_movies['listed_in'].str.contains('&'.join(most_common_genres.index))]['title'].iloc[0]

# Verileri birleştirme
result = pd.concat([movies_per_year], axis=1)
result.columns = ['Movies']

# En yüksek puanlı filmi sonuçlara ekleme
result['Highest Rated Movie'] = highest_rated_movie

# Grafik oluşturma
result.plot(kind='bar', figsize=(12,6), color=['green', 'purple'])


```

```python

#hangi kriterlerid
import pandas as pd

class Film:
    def __init__(self, title, director, release_year, type):
        self.title = title
        self.director = director
        self.release_year = release_year
        self.type = type

    def __repr__(self):
        return f"{self.title} ({self.release_year}) - {self.type}"

class FilmListesi:
    def __init__(self, dosya_adi):
        self.dosya_adi = dosya_adi
        self.filmler = []

    def dosya_oku(self):
        try:
            df = pd.read_csv(self.dosya_adi)
            for _, row in df.iterrows():
                film = Film(row['title'], row['director'], row['release_year'], row['type'])
                self.filmler.append(film)
            print("Veri dosyası okundu.")
        except FileNotFoundError:
            print("Dosya bulunamadı, lütfen geçerli bir dosya adı girin.")
        except Exception as e:
            print("Bir hata oluştu:", e)
            exit()

    def filmleri_listele(self):
        if len(self.filmler) == 0:
            print("Listelenecek film yok.")
        else:
            for film in self.filmler:
                print(film)

    def filmleri_filtrele(self, tip):
        filtrelenmis_filmler = [film for film in self.filmler if film.type == tip]
        if len(filtrelenmis_filmler) == 0:
            print("Aradığınız kriterlere uygun film bulunamadı.")
        else:
            for film in filtrelenmis_filmler:
                print(film)

    def eksik_deger_kontrolu(self):
        df = pd.read_csv(self.dosya_adi)
        if df.isnull().values.any():
            print("Veri setinde eksik değerler var.")
            print(df.isnull().sum())
        else:
            print("Veri setinde eksik değer yok.")

# Kullanım örneği
film_listesi = FilmListesi('netflix_titles.csv')
film_listesi.dosya_oku()
film_listesi.filmleri_listele()
film_listesi.filmleri_filtrele("Movie")
film_listesi.eksik_deger_kontrolu()
```

```python
import pandas as pd

# Veri setini yükleme
df = pd.read_csv("netflix_titles.csv")

# Kullanıcıdan ülke ismini girdi olarak alma
country = input("Ülke ismini girin: ")

# Ülkeye göre hangi kategorilerde içeriklerin olduğunu gösteren bir fonksiyon
def show_categories(country):
    categories = []
    for index, row in df.iterrows():
        if str(row["country"]).lower() == str(country).lower():
            for category in row["listed_in"].split(", "):
                if category not in categories:
                    categories.append(category)
    categories = sorted(categories)
    print(f"{country} için mevcut kategoriler:")
    for category in categories:
        print("- " + category)

# Fonksiyonu çağırma
show_categories(country)

# Kullanıcıya belirli bir kategoriye göre içerik öneren bir fonksiyon
def recommend_content_by_category(category):
    filtered_data = df[df["listed_in"].str.contains(category, case=False)]
    sorted_data = filtered_data.sort_values("release_year", ascending=False)
    return sorted_data.head(10)

# Karar kontrolü ile hangi kategoriye göre öneri yapılacağına karar verme
while True:
    categories = sorted(list(set(df["listed_in"].str.cat(sep=", ").split(", "))))
    answer = input("Hangi kategoriye göre öneri yapmak istersiniz? (Çıkmak için 'q' ya basın)\n")
    if answer.lower() == "q":
        break
    elif answer in categories:
        recommendations = recommend_content_by_category(answer)
        print(f"{answer} kategorisine ait öneriler:")
        for index, row in recommendations.iterrows():
            country = row["country"]
            if pd.isna(country):
                country = "Bilinmiyor"
            print(f"{row['title']} ({row['release_year']}) - {country}")
    else:
        print("Geçersiz bir seçim yaptınız. Lütfen listeden bir kategori seçin.")
```

```python
 #TR  kategorilerini listeleme ve kullanıcının seçtiği bir kategoriye göre içerik önerme
 
import pandas as pd

df = pd.read_csv('netflix_titles.csv')


def show_categories(country):
    categories = []
    for index, row in df.iterrows():
        if str(row["country"]).lower() == str(country).lower():
            for category in row["listed_in"].split(", "):
                categories.append(category)
    categories = sorted(list(set(categories)))
    print("Bu ülkede mevcut olan kategoriler:")
    for category in categories:
        print("- " + category)



show_categories("Turkey")

def recommend_content_by_category(category):
    filtered_data = df[df["listed_in"].str.contains(category)]
    sorted_data = filtered_data.sort_values("release_year", ascending=False)
    return sorted_data[["title", "country", "release_year"]].head(10)


category = input("Bir kategori girin: ")
recommendations = recommend_content_by_category(category)

# Önerileri ekrana yazdırma
print("Bu kategoriye ait öneriler:")
for index, row in recommendations.iterrows():
    country = row["country"]
    if pd.isna(country):
        country = "Bilinmiyor"
    print("- " + row["title"] + " (" + str(row["release_year"]) + ") - " + country)

#classs hariç hepsi var
```
