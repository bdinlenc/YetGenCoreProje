```python

#öneri (girdi aldık ,karar yapısı yaptık)
import random
import pandas as pd

# Netflix veri setini yükleme
df = pd.read_csv('netflix_titles.csv')

# Kullanıcıdan seçim yapmasını isteme
choice = input("Dizi mi yoksa film mi önerilsin? ")

# Sadece filmleri filtrelemek
if choice == "film":
    df_movies = df[df['type'] == 'Movie']
    if not df_movies.empty:
        # Rastgele bir film seçmek
        suggestion = df_movies.sample()
        print("Önerimiz: ")
        print("Film adı:", suggestion['title'].values[0])
        print("Türü:", suggestion['listed_in'].values[0])
        print("Yapım yılı:", suggestion['release_year'].values[0])
        
    else:
        print("Aradığınız kriterlere uygun film bulunamadı.")
        
# Sadece dizileri filtrelemek
elif choice == "dizi":
    df_tv = df[df['type'] == 'TV Show']
    if len(df_tv) > 0:
        # Rastgele bir dizi seçmek
        suggestion = df_tv.sample()
        print("Önerimiz: ")
        print("Dizi adı:", suggestion['title'].values[0])
        print("Türü:", suggestion['listed_in'].values[0])
        print("Yapım yılı:", suggestion['release_year'].values[0])
        
    else:
        print("Aradığınız kriterlere uygun dizi bulunamadı.")
        
# Geçersiz seçim durumu
else:
    print("Lütfen geçerli bir seçim yapın: 'dizi' veya 'film'.")
```

```python
#hata var 
import pandas as pd

# Verileri yükleme
df = pd.read_csv('netflix_titles.csv')

# Sadece film ve TV şovlarını içeren veri setleri oluşturma
df_movies = df[df['type'] == 'Movie']
df_tv_shows = df[df['type'] == 'TV Show']

# Yıla göre gruplayıp toplam sayıları hesaplama
movies_per_year = df_movies.groupby('release_year').size()
tv_shows_per_year = df_tv_shows.groupby('release_year').size()

# En yüksek puanlı filmi bulma ve adını getirme
highest_rated_movies = df_movies.loc[df_movies.groupby('release_year')['rating'].idxmax()][['release_year', 'title']]
highest_rated_movies.set_index('release_year', inplace=True)

# Verileri birleştirme
result = pd.concat([movies_per_year, tv_shows_per_year, highest_rated_movies], axis=1)
result.columns = ['Movies', 'TV Shows', 'Highest Rated Movie']

# Grafik oluşturma
result.plot(kind='bar', figsize=(12,6), color=['green', 'blue', 'purple'])
```

```python

#lidere sor

import pandas as pd

# Veri setini oku
df = pd.read_csv('netflix_titles.csv')

# Rating verilerini string olarak tanımla
df['rating'] = df['rating'].astype(str)

# Fonksiyonu tanımla
def recommend_content(data, content_type, genre, country):
    # İstenilen kriterlere göre filtreleme yap
    filtered_data = data[(data['type'] == content_type) & 
                         (data['listed_in'].str.contains(genre)) & 
                         (data['country'].str.contains(country))]

    # Rating verilerini string olarak tanımladığımızdan dolayı burada string metotlarını kullanabiliriz
    sorted_data = filtered_data.sort_values('rating', ascending=False)

    # İlk 5 öneriyi döndür
    return sorted_data[['title', 'rating']].head(5)

# Fonksiyonu kullanarak önerileri al ve ekrana yazdır
recommendations = recommend_content(df, content_type='TV Show', genre='Documentaries', country='Turkey')
print(recommendations)


```

```python
#class yapısı var içeride dögü var minik ,fonksiyon da var


import pandas as pd

class Film:
    def __init__(self, title, director, release_year, type):
        self.title = title
        self.director = director
        self.release_year = release_year
        self.type = type

    def __repr__(self):
        return f"{self.title} ({self.release_year}) - {self.type}"

# CSV dosyasını bir DataFrame olarak okuyun
df = pd.read_csv('netflix_titles.csv')

# DataFrame sütunlarından uygun verileri alarak Film sınıfına ait nesneleri oluşturun
filmler = []
for _, row in df.iterrows():
    film = Film(row['title'], row['director'], row['release_year'], row['type'])
    filmler.append(film)

# Filmleri yazdırın
for film in filmler:
    if film.type == "Movie":
        print(film)

 
```
