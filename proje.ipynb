```python
import pandas as pd

# Veri setini oku
df = pd.read_csv('netflix_titles.csv')

# "listed_in" sütunundaki her öğeyi virgülle ayırarak bir liste oluştur
genre_lists = df['listed_in'].str.split(', ')

# Tüm türleri tek bir listede topla
all_genres = []
for genre_list in genre_lists:
    all_genres += genre_list
    
# Her türün sayısını say
genre_counts = pd.Series(all_genres).value_counts()

# En çok izlenen 10 türü yazdır
print(genre_counts.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')

# 'listed_in' sütunundaki her bir türün sayısını hesaplayalım
genre_counts = df['listed_in'].value_counts().head(10)

# Bar grafiğini çizelim
plt.bar(genre_counts.index, genre_counts.values)
plt.xticks(rotation=90)
plt.title('En Çok İzlenen Türler')
plt.xlabel('Tür')
plt.ylabel('Sayı')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')

df = df.dropna()  # Eksik değerleri çıkar

df_tr = df[df['country'].str.contains('Turkey')]  # Türkiye'deki içeriklerin filtrelenmesi

top_genres = df_tr['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)

import matplotlib.pyplot as plt #daire grafiği çizilmesi

plt.pie(top_genres.values, labels=top_genres.index)

plt.show()

#trde en cok izlenen turlerin daire grafiği
```
